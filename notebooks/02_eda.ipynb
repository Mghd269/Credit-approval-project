### Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Assigning data into dataframes

```python
df1 = pd.read_csv('data/02_interim/categorical_data.csv')
df2 = pd.read_csv('data/02_interim/numerical_data.csv')

print(df1.head())
print(df2.head())
```

### Statistics

```python
print(df2.describe().round(2))
```

```python
sns.heatmap(df2.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 20))

ax = sns.countplot(y=df1['Profession'])

plt.title("Frequency of each Profession")
plt.ylabel("Profession")
plt.xlabel("Count")

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
count = df1['Married/Single'].value_counts()

plt.pie(count, labels=count.index, autopct='%1.1f%%', startangle=90, colors=['orange', 'green'])
plt.title('Marriage Status')
plt.show()
```

```python
count = df1['House_Ownership'].value_counts()

plt.pie(count, labels=count.index, autopct='%1.1f%%')
plt.title('House Ownership')
plt.show()
```

```python
count = df1['Risk_Flag'].value_counts()
labels = ['approved' if x == 0 else 'reject' for x in count.index]

plt.pie(count, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Credit approval %')

plt.show()
```

```python
uniqe_city = df1['CITY'].nunique()
print("Number of unique cities: ", uniqe_city)

uniqe_state = df1['STATE'].nunique()
print("Number of unique states: ", uniqe_state)
```

```python
ax2 = sns.countplot(y=df1['STATE'])
for container in ax2.containers:
    ax2.bar_label(container)

plt.title("Frequency of each State")    
plt.show()
```
